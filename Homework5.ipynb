```python
#W6A16: Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣
#trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải
try:
  n, m = map(int, input("Nhập số hàng và số cột: ").strip().split())
  i = 0
  matrix = []
  while i < n:
    hang = list(map(int, input(f"Nhập hàng thứ {i+1}: ").strip().split()))
    if len(hang) == m:
      matrix.append(hang)
      i += 1
    else:
      print("Đầu vào không hợp lệ, hãy nhập lại!")
  for i in range(n):
    for j in range(m):
      print(f"{matrix[i][j]: 4d}", end = '')
    print()
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
#W6A17: Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính
#và đường chéo phụ của mảng
try:
  n = int(input("Nhập số hàng: "))
  i = 0
  matrix = []
  while i < n:
    hang = list(map(int, input(f"Nhập hàng thứ {i+1}: ").strip().split()))
    if len(hang) == n:
      matrix.append(hang)
      i += 1
    else:
      print("Đầu vào không hợp lệ, hãy nhập lại!")
  print("Đường chéo chính:")
  for a in range(n):
     print(matrix[a][a], end=" ")
  print()
  print("Đường chéo phụ:")
  for b in range(n):
     print(matrix[b][n - 1 - b], end=" ")
  print()
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
#W6A18: Nhập vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k
#nhập trên cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các
#phần tử trên cột k (hàng k)
try:
  n, m, k = map(int, input("Nhập số hàng, số cột và số nguyên k: ").strip().split())
  if k > n or k > m or k < 1:
    print("Đầu vào không hợp lệ!!!")
    exit()
  i = 0
  matrix = []
  while i < n:
    hang = list(map(int, input(f"Nhập hàng thứ {i+1}: ").strip().split()))
    if len(hang) == m:
      matrix.append(hang)
      i += 1
    else:
      print("Đầu vào không hợp lệ, hãy nhập lại!")
  th = sum(matrix[k-1])
  print(f"Tổng hàng thứ {k} là: {th}" )
  tc = sum(matrix[i][k-1] for i in range(n))
  print(f"Tổng cột thứ {k} là: {tc}" )
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
#W6A20: Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy
#tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu
#không có thì trả về -1
def vtri(lst, k):
 if k in lst:
  return lst.index(k)
 else:
  return -1
try:
  n = list(map(int, input("Nhập dãy số nguyên: ").strip().split()))
  k = int(input("Nhập số k: "))
  print(vtri(n, k))
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
#W6A21: Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary
#ánh xạ chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện
try:
  n = input("Nhập các chuỗi: ").strip().split()
  dic = {}
  for i in set(n):
    tup = ()
    for j in range(len(n)):
      if n[j] == i:
        tup = tup + (j,)
    dic[i] = tup
  print(dic)
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
#W6A22: Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá
#trị lớn nhất trong list
def tim(lst):
  num_max = max(lst)
  return [i for i, v in enumerate(lst) if v == num_max]
try:
  n = list(map(int, input("Nhập dãy số nguyên: ").strip().split()))
  lst = tim(n)
  print("giá trị lớn nhất ở vị trí ", *lst)
except ValueError:
  print('Đầu vào không hợp lệ!!!')
```

```python
#W6A23: VIết hàm xóa đi các số <x có trong mảng
def xoa(lst, k):
  return [i for i in lst if i >= k]
try:
  n = list(map(int, input("Nhập dãy số nguyên: ").strip().split()))
  k = int(input("Nhập số nguyên k: "))
  print(xoa(n, k))
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
#W6A24: Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn
#đúng vị trí trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9
def dao(lst):
    zeros = [x for x in lst if x == 0]
    others = [x for x in lst if x != 0]
    return zeros + others
try:
  n = list(map(int, input("Nhập dãy số nguyên: ").strip().split()))
  print(dao(n))
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
#W6A26: Tính tổng các phần tử cực đại địa phương trong một danh sách,
#không tính 2 phần tử đầu và cuối
try:
  n = list(map(int, input("Nhập dãy số nguyên: ").strip().split()))
  lst = [n[i] for i in range(1, len(n)-1) if (n[i-1] < n[i] and n[i+1] < n[i])]
  print("Tổng các phân tử cực đại địa phương: ", sum(lst))
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
#W6A27: Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá
#trị cho trước). nếu không có in ra -1
try:
  n = list(map(int, input("Nhập dãy số nguyên: ").strip().split()))
  x = int(input("Nhập số nguyên x: "))
  lst =  [i for i in range(len(n)) if n[i] < x]
  print(f"Vị trí các số < {x}: ", *lst)
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
#W6A28: Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
#thực sự, hay vô danh (không tăng, không giảm)
try:
  n = list(map(int, input("Nhập dãy số nguyên: ").strip().split()))
  tang, giam = True, True
  for i in range(1, len(n)):
    if n[i] > n[i-1]:
      giam = False
    elif n[i] < n[i-1]:
      tang = False
  if giam:
    print("Dãy giảm thật sự!!")
  elif tang:
    print("Dãy tăng thật sự!!")
  else:
    print("Dãy vô danh!!")
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```
