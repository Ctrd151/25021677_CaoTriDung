```python
#Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể là
#độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'.
#Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường.
#Tính diện tích, chu vi của tam giác và in ra màn hình.
a,b,c=map(float,input().split())
if a+b>c and a+c>b and b+c>a:
  if a==b and b==c:
    print('là tam giác đều')
  elif a==b or b==c or a==c:
    print('là tam giác cân')
  elif a*a+b*b==c*c or b*b+c*c==a*a or a*a+c*c==b*b:
    print('là tam giác vuông')
  else:
    print('là tam giác thường')
  p=(a+b+c)/2
  print(f'chu vi: {p*2} diện tích: {(p*(p-a)*(p-b)*(p-c))**(1/2)}')
else:
  print('Khong la ba canh cua mot tam giac')
```

```python
#Bài 6: Tìm giá trị lớn nhất trong 4 số a, b, c, d (a, b, c, d được nhập từ bàn phím).
#Bài 7: Tìm giá trị nhỏ nhất trong 4 số a, b, c, d (a, b, c, d được nhập từ bàn phím).
l = list(map(float, input('Nhập 4 số a, b, c, d: ').split()))
max=l[1]
min=l[1]
for i in range(0,len(l)-1):
  if l[i+1]>max:
    max=l[i+1]
  elif l[i+1]<min:
    min=l[i+1]
print(f'giá trị lớn nhất là {max}, giá trị nhỏ nhất là {min}')
```

```python
#Bài 19: Nhập vào một số nguyên không âm, kiểm tra xem nó có phải là số nguyên tố hay không?
n = int(input("Nhập số nguyên n không âm: "))

if n < 0:
    print(f"{n} là số nguyên âm")
elif n < 2:
    print(f"{n} không phải số nguyên tố")
else:
    la_nguyen_to = True
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            la_nguyen_to = False
            break
    if la_nguyen_to:
        print(f"{n} là số nguyên tố")
    else:
        print(f"{n} không phải số nguyên tố")
```

```python
#Bài 20: In ra các số nguyên tố nhỏ hơn hoặc bằng N (N là số nguyên không âm được nhập từ bàn phím).
n=int(input('nhập số nguyên n: '))
for x in range(2,n+1):
    la_nguyen_to = True
    for i in range(2, int(x**0.5) + 1):
        if x % i == 0:
            la_nguyen_to = False
            break
    if la_nguyen_to:
      print(x)
```

```python
#Bài 21: Nhập 3 loại tiền và số tiền cần đổi. Hãy tìm tất cả các tổ hợp có được của 3 loại tiền trên cho số tiền vừa nhập.
a = int(input("Nhập mệnh giá tiền thứ 1: "))
b = int(input("Nhập mệnh giá tiền thứ 2: "))
c = int(input("Nhập mệnh giá tiền thứ 3: "))
n = int(input("Nhập số tiền cần đổi: "))
count = 0
for x in range(n // a + 1):
    for y in range(n // b + 1):
        for z in range(n // c + 1):
            if a*x + b*y + c*z == n:
                count += 1
                print(f"{count}. {x} tờ {a}, {y} tờ {b}, {z} tờ {c}")
if count == 0:
    print("Không có tổ hợp nào phù hợp.")
else:
    print(f"Tổng cộng có {count} tổ hợp.")
```
