```python
#W5A1:Viêt hàm truyền vào hai số nguyên a, b. Trả vế số lớn hơn
def Max_Of_Two(a,b):
  if a > b:
    return a
  else:
    return b
try:
  a, b = map(int,input("Nhập hai số nguyên a, b: ").split())
  print(f"Số lớn nhất là {Max_Of_Two(a, b)}")
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```

```python
#W5A3: Viết hàm kiểm tra một số có phải là số nguyên tố hay không?
def prime(a):
  if a < 2:
    return False
  for i in range(2, int(a**0.5)+1):
    if a % i == 0:
      return False
  return True
try:
  a = int(input("Nhập số nguyên dương a: "))
  if a >= 2:
    if prime(a):
      print(f"{a} là số nguyên tố")
    else:
      print(f"{a} không là số nguyên tố")
  else:
    print("!!!Đầu vào không hợp lệ")
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```

```python
#W5A4: VIết hàm kiểm tra một số có phải số hoàn hảo hay không,
#xử lý trên dữ liệu kiểu số nguyên
def perfect_num(a):
  tong = 0
  for i in range(1, int(a**0.5)+1):
    if a % i == 0:
      tong += i
      if a/i != i:
        tong += a/i
  if 2*a == tong:
    return True
  else:
    return False
try:
  a = int(input("Nhập số nguyên dương a: "))
  if a > 0:
    if perfect_num(a):
      print(f"{a} là số hoàn hảo")
    else:
      print(f"{a} không là số hoàn hảo")
  else:
    print("!!!Đầu vào không hợp lệ")
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```

```python
#W5A5: Viết hàm đưa vào một list số nguyên và một số nguyên dương k
#Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong
#list số nguyên, nếu không có thì trả về -1
def search(k, lst):
  for i in range(len(lst)):
    if lst[i] == k:
      return i
  return -1
try:
  lst = list(map(int, input("Nhập dãy số nguyên: ").split()))
  k = int(input("Nhập số nguyên k: "))
  a = search(k, lst)
  if a == -1:
    print(f"Không tìm thấy {k} trong list")
  else:
    print(f"Giá trị {k} đầu tiên ở vị trí {a}")
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```

```python
#W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n,
#trả về giai thừa của n (n!)
def factorial(n):
    fac = 1
    for i in range(1, n+1):
      fac *= i
    return fac
try:
  n = int(input("Nhập số nguyên n: "))
  if n > 0:
    print(f"{n}! = {factorial(n)}")
  else:
    print("!!!Đầu vào không hợp lệ")
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```

```python
#W5A7: viết hàm giả lập máy tính bỏ túi
def tinh(num1, dau, num2):
  if dau == "+":
    return num1 + num2
  if dau == "-":
    return num1 - num2
  if dau == "*":
    return num1 * num2
  if dau == "/":
    return num1 / num2
try:
  num1 = float(input("Nhập số thực thứ nhất: "))
  dau = input("Nhập dấu: ")
  num2 = float(input("Nhập số thực thứ hai: "))
  print(f"{num1} {dau} {num2} = {tinh(num1, dau, num2)}")
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```

```python
#W5A8: Tính khoảng cách Hamming
def hamming(x, y):
  binx, biny = "", ""
  while x != 0:
    binx = str(x%2) + binx
    x //= 2
  while y != 0:
    biny = str(y%2) + biny
    y //= 2
  count = 0
  lon = max(len(binx), len(biny))
  nho = min(len(binx), len(biny))
  if len(binx) < len(biny):
    binx = (lon - nho)*"0" + binx
  elif len(binx) > len(biny):
    biny = (lon - nho)*"0" + biny
  for i in range(lon):
    if binx[i] != biny[i]:
      count += 1
  return count
try:
  a, b = map(int,input("Nhập hai số nguyên a, b: ").split())
  print(f"Khoảng cách Hamming giữa {a} và {b} là {hamming(a, b)}")
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```

```python
#W5A9 Viết hàm trả về tổng các chữ số của số nguyên dương n
def tong(n):
  z = 0
  for i in str(n):
    z += int(i)
  return z
try:
  n = int(input("Nhập số nguyên n: "))
  if n > 0:
    print(f"Tổng các chữ số của {n} là {tong(n)}")
  else:
    print("!!!Đầu vào không hợp lệ")
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```

```python
#W5A10: kiểm tra hai từ là đẳng cấu
def dang_cau(x, y):
  lst = []
  for i in range(len(x)):
    lst.append([x[i], y[i]])
  for i in range(len(x)):
    for j in range(len(x)):
      if lst[i][0] == lst[j][0]:
       if lst[i][1] != lst[j][1]:
        return False
      if lst[i][1] == lst[j][1]:
        if lst[i][0] != lst[j][0]:
          return False
  return True
try:
  x = input('Nhập từ thứ nhất: ')
  y = input("Nhập từ thứ hai: ")
  if len(x) == len(y):
    if dang_cau(x, y):
      print(f"{x} và {y} là hai từ đẳng cấu")
    else:
      print(f"{x} và {y} là không hai từ đẳng cấu")
  else:
    print("!!!Đầu vào không hợp lệ")
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```
