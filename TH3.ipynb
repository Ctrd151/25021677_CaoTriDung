```python
#W3A9
#Nhập vào 3 số nguyên dương a, b, c. Kiểm tra xem a, b, c có cấu thành độ
#dài cạnh của 1 tam giác được không. Nếu a, b, c cấu tạo thành được một tam
#giác, in ra "Yes". Ngược lại in ra "No".
try:
  a,b,c=map(int,input("nhập độ dài 3 cạnh: ").split())
  if a>0 and b>0 and c>0:
    if a+b>c and a+c>b and b+c>a:
      print('"Yes"')
    else:
      print('"No"')
  else:
    print('Lỗi: nhập giá trị chưa đúng!')
except ValueError:
  print('Lỗi: nhập giá trị chưa đúng!')
```

```python
#W3A10
#Nhập vào 4 số nguyên. In ra màn hình số nguyên lớn nhất trong 4 số.
try:
  lst=list(input('nhập 4 số nguyên: ').split())
  if len(lst)==4:
    max=lst[0]
    for i in range(len(lst)):
      if lst[i]>max:
        max=lst[i]
    print(max)
  else:
    print('lỗi: nhập không đúng 4 số nguyên')
except ValueError:
  print('Lỗi: nhập không đúng 4 số nguyên')
```

```python
#W3A11
#Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm
#tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không. Nếu không in
#ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều",
#"Tam giác cân", hay "Tam giác thường".

a,b,c=map(float,input().split())
if a+b>c and a+c>b and b+c>a:
  if a==b and b==c:
    print('Tam giác đều')
  elif a==b or b==c or a==c:
    print('Tam giác cân')
  elif a*a+b*b==c*c or b*b+c*c==a*a or a*a+c*c==b*b:
    print('Tam giác vuông')
  else:
    print('Tam giác thường')
else:
  print('Khong la ba canh cua mot tam giac')
```

```python
#W3A15
#[Xếp loại học lực]
#Nhập vào điểm trung bình của một học sinh và in ra học lực của học sinh đó.
#Xếp loại học lực theo quy tắc:
#- >= 8.0: Giỏi
#- >= 6.5: Khá
#- >= 5.0: Trung bình
#- < 5.0: Yếu
try:
    diem_tb = float(input("Nhập điểm trung bình của học sinh: "))
    if diem_tb < 0 or diem_tb > 10:
        print("Lỗi: điểm phải nằm trong khoảng từ 0 đến 10.")
    else:
        if diem_tb >= 8.0:
            print("Học lực: Giỏi")
        elif diem_tb >= 6.5:
            print("Học lực: Khá")
        elif diem_tb >= 5.0:
            print("Học lực: Trung bình")
        else:
            print("Học lực: Yếu")
except ValueError:
    print("Lỗi: Vui lòng nhập số hợp lệ!")
```

```python
#W3A16
#[Rounding]
#Nhập vào một số thực: in ra kết quả làm tròn lên, tròn xuống, làm tròn tới số
#nguyên gần nhất (không được sử dụng hàm sẵn có của pyton). Các kết quả
#cách nhau một dấu cách
n=float(input())
if n%1<0.5:
  x=n//1
else:
  x=n//1+1
print(f'{n//1}, {n//1+1}, {x}')
```
