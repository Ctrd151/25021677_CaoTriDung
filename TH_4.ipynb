```python
#W4A8
#ReverseAddPalindrome]
#Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
#đối xứng (palindrome). In ra số bước và giá trị palindrome thu
def reverse_number(num):
  rev=0
  while num>0:
    rev=rev*10+num%10
    num //=10
  return rev
try:
 n=int(input('nhập số nguyên dương n: '))
 if n>0:
  step=0
  while True:
    rev=reverse_number(n)
    if n==rev:
      break
    n += rev
    step += 1
  print(f'số bước là {step}, giá trị palindrome là {n}')
 else:
  print('Lỗi: chưa nhập số nguyên dương')
except ValueError:
  print('Lỗi: nhập chưa đúng giá trị')
```

```python
#W4A10
#"[LongestCollatzUpToN]
#Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
#và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
#nhỏ nhất."
def collatz(n):
  step=0
  while n!=1:
    if n%2==0:
      n//=2
    else:
      n=n*3+1
    step+=1
  return step
try:
  n=int(input('nhập số nguyên dương n: '))
  if n>0:
    num=0
    max_step=0
    for i in range(1,n+1):
      steps=collatz(i)
      if steps>max_step:
        num=i
        max_step=steps
    print(num,max_step)
  else:
    print('Lỗi: chưa nhập giá trị đúng')
except ValueError:
  print('Lỗi: nhập giá trị chưa đúng')
```
