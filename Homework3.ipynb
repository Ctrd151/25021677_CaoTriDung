```python
#Bài 3:Nhập vào số nguyên dương a, in ra bảng cửu chương của a
try:
  a = int(input("Nhập số nguyên dương a: "))
  if 0 < a < 10:
    print(f"Bảng cửu chương của {a}:")
    for i in range(1,11):
      print(f"{a} x {i} = {a*i}")
  else:
    print("!!!Đầu vào không hợp lệ")
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```

```python
#Bài 4: Viết chương trình in ra hình tam giác cân có độ cao
#h được nhập từ bàn phím
try:
  h = int(input("Nhập độ cao h: "))
  if h > 0:
    for i in range(h):
      if i == 0:
          print(" "*(h-i-1) + "*")
      elif i == h-1:
          print("*"*(h*2-1))
      else:
          print(" "*(h-i-1) + "*" + " "*(2*i-1) + "*")
  else:
    print("Đầu vào không hợp lệ!!!")
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
#Bài 5: Nhập vào số nguyên dương n. Tính S = 1 + 2 + 3 + 4 + … + n
try:
  n = int(input("Nhập số nguyên dương n: "))
  if n > 0:
    tong = 0
    for i in range(1,n+1):
      tong += i
    print(f"Tổng dãy S = {tong}")
  else:
    print("!!!Đầu vào không hợp lệ")
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```

```python
#Bài 6: Nhập vào số nguyên dương a, in ra tất cả các ước dương của a
try:
  a = int(input("Nhập số nguyên dương a: "))
  if a > 0:
    uoc = []
    for i in range(1, int(a**0.5)+1):
      if a % i == 0:
        uoc.append(i)
        uoc.append(a/i)
    lst = sorted(set(uoc))
    for n in lst:
      print(int(n))
  else:
    print("!!!Đầu vào không hợp lệ")
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```

```python
#Bài 7: Nhập vào số nguyên dương a, đếm số ước dương của a
try:
  a = int(input("Nhập số nguyên dương a: "))
  if a > 0:
    uoc = []
    for i in range(1, int(a**0.5)+1):
      if a % i == 0:
        uoc.append(i)
        uoc.append(a/i)
    uoc = set(uoc)
    print(len(uoc))
  else:
    print("!!!Đầu vào không hợp lệ")
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```

```python
#Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b
try:
  a, b = map(int,input("Nhập số nguyên dương a, b: ").split())
  if a > 0 and b > 0:
    uoc_chung = []
    for i in range(1, min(a, b)+1):
      if a % i == 0 and b % i == 0:
        uoc_chung.append(i)
    print(*uoc_chung)
  else:
    print("!!!Đầu vào không hợp lệ")
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```

```python
#Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số
#hoàn hảo hay không. Số hoàn hảo là số có tổng các ước
#(ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28…
try:
  a = int(input("Nhập số nguyên dương a: "))
  if a > 0:
    uoc = []
    for i in range(1, int(a**0.5)+1):
      if a % i == 0:
        uoc.append(i)
        uoc.append(a/i)
    uoc = set(uoc)
    tong = 0
    for n in uoc:
      tong += n
    if 2*a == tong:
      print(f"{n} là số hoàn hảo")
    else:
      print(f"{n} không phải số hoàn hảo")
  else:
    print("!!!Đầu vào không hợp lệ")
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```

```python
#Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1.
#Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập.
#KHÔNG ĐƯỢC dùng mảng
try:
  a = int(input("Nhập 1 số nguyên : "))
  max = a
  min = a
  while a != -1:
    a = int(input("Nhập 1 số nguyên : "))
    if a > max:
      max = a
    elif a < min:
      min = a
  print(f"Số lớn nhất là {max}, số bé nhất là {min}")
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
#Bài 15: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số.
#Làm bằng nhiều cách nhất có thể, ko dùng mảng.
try:
  n = abs(int(input("Nhập số nguyên n : ")))
  count = 0
  while n != 0:
    n = n // 10
    count += 1
  print(f"Có {count} chữ số")
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
#Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số
#chẵn, bao nhiêu chữ số lẻ
try:
  n = input("Nhập số nguyên dương n: ")
  if int(n) > 0:
    count_even = 0
    count_odd = 0
    for i in n:
      if int(i) % 2 == 0:
        count_even += 1
      else:
        count_odd +=1
    print(f"Có {count_even} chữ số chẵn và {count_odd} chữ số lẻ")
  else:
    print("Đầu vào không hợp lệ!!!")
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```

```python
#Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1,
#2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó.
#Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci
#lớn nhất nhưng không vượt quá A
try:
  a = int(input("Nhập số nguyên dương A: "))
  if a > 0:
    f1, f2 = 1, 1
    while f2 <= a:
      f1, f2 = f2, f1 + f2
    print(f"Số fibonacci lớn nhất nhưng không vượt qua {a} là {f1}")
  else:
    print("!!!Đầu vào không hợp lệ")
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```
