```python
#W6A1: Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng
#giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
try:
  n = list(map(int, input("Nhập một dòng gồm các số nguyên: ").split()))
  da_gap = []
  for i in n:
    if i not in da_gap:
      da_gap.append(i)
  print(da_gap)
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
#W6A2: Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ
#đầu tới vị trí i), giữ nguyên thứ tự
try:
  n = list(map(int, input("Nhập một dãy số nguyên: ").split()))
  ttl = []
  tong = 0
  for i in n:
    tong += i
    ttl.append(tong)
  print(ttl)
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
#W6A3: Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra
#tuple sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay;
#không sắp xếp.)
try:
  n = tuple(map(int, input("Nhập một dãy số nguyên: ").split()))
  k = int(input("Nhập số nguyên k: "))
  if 0 <= k <= len(n):
    a = n[k:] + n[:k]
    print(a)
  else:
    print("Đầu vào không hợp lệ!!!")
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
#W6A4: Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
#dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key
#lặp lại, thêm vào list của key đó
try:
  n = input("Nhập dãy các cặp key:value: ").split()
  dic = {}
  for i in n:
    key, value = i.split(":")
    value = int(value)
    if key not in dic:
       dic[key] = []
    dic[key].append(value)
  print(dic)
except:
  print("Đầu vào không hợp lệ!!!")
```

```python
#W6A5: Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ...,
#'zeros': ...} (đếm số >0, <0, =0)
try:
  n = list(map(int, input("Nhập một dãy số nguyên: ").split()))
  dic = {'positives':0, 'negatives':0, 'zero':0}
  for i in n:
    if i < 0:
      dic['positives'] += 1
    elif i > 0:
      dic['negatives'] += 1
    else:
      dic['zero'] += 1
  print(dic)
except ValueError:
  print('Đầu vào không hợp lệ!!!')
```

```python
#W6A8: Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất
#hiện của từng chuỗi và lưu vào dictionary
try:
  n = list(map(str, input("Nhập dãy các chuỗi: ").strip().split()))
  dic = {}
  for i in n:
    if i not in dic.keys():
      count = 0
      for j  in n:
        if i == j:
          count += 1
      dic[i] = count
  print(dic)
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```

```python
#W6A9: Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một
#dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ
#được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới
#theo thứ tự từ điển tăng dần của key
def nhap_dic(n):
    dic = {}
    for i in n:
      key, value = i.split(":")
      value = int(value)
      key = int(key)
      dic[key] = value
    return dic
try:
  n1 = input("Nhập dictionary 1: ").strip().split()
  n2 = input("Nhập dictionary 2: ").strip().split()
  dic1, dic2 = nhap_dic(n1), nhap_dic(n2)
  dic3 = {}
  for k in dic1.keys() | dic2.keys():
    dic3[k] = dic1.get(k, 0) + dic2.get(k, 0)
  dic3 = dic1 | dic2
  dic3_sorted = dict(sorted(dic3.items()))
  print(dic3_sorted)
except ValueError:
  print("!!!Đầu vào không hợp lệ")
```

```python
'''
W6A10: Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên
dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là
một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh
sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được
sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.
'''
try:
  n = list(map(int, input("Nhập danh sách các số nguyên: ").strip().split()))
  k = int(input("Nhập số k: "))
  tup = []
  for i in range(len(n)):
    for j in range(i + 1, len(n)):
      if n[i] + n[j] == k:
       tup.append((n[i], n[j]))
  tup.sort(key=lambda x: x[0])
  print(tup)
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
'''
W6A11:[Tách số chẵn và lẻ từ tuple]
Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa
tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên
thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt
'''
try:
  n = tuple(map(int, input("Nhập dãy số nguyên: ").strip().split()))
  even = ()
  odd = ()
  for i in n:
    if i % 2 == 0:
      even = even + (i,)
    else:
      odd = odd + (i,)
  print("Dãy chẵn: ", even)
  print("Dãy lẻ: ", odd)
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
'''
W6A12: [Tìm phần tử xuất hiện nhiều nhất]
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong
list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ
nhất trong số đó.
'''
try:
  n = list(map(int, input("Nhập dãy số nguyên: ").strip().split()))
  max_count = 0
  val = None
  n1 = set(n)
  for i in n1:
    count = 0
    for j in n:
      if i == j:
        count += 1
    if count > max_count:
      max_count = count
      val = i
    if count == max_count:
      if i < val:
        val = i
  print(f"Giá trị {val} xuất hiện nhiều nhất với {max_count} lần")
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
'''
W6A13: [Đảo ngược key và value của dictionary]
Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo
một dictionary mới bằng cách đảo ngược key và value của dictionary ban
đầu. In ra dictionary mới.
'''
try:
  n = input("Nhập key:value : ").strip().split()
  dic = {}
  for item in n:
    key, value = item.split(':')
    value = int(value)
    dic[key] = value
  print(dic)
  new_dic = {}
  for k, v in dic.items():
    new_dic[v] = k
  print(new_dic)
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
'''
W6A14: [Giao của hai danh sách]
Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm các
phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các
phần tử chung đó, không trùng lặp.
'''
try:
  n = list(map(int, input("Nhập dãy số nguyên 1: ").strip().split()))
  m = list(map(int, input("Nhập dãy số nguyên 2: ").strip().split()))
  print(f"Giao của hai danh sách là: {[x for x in (set(n) & set(m))]}")
except ValueError:
  print("Đầu vaog không hợp lệ!!!")
```

```python
'''
W6A15: [Lọc dictionary theo giá trị]
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp
key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary
mới.
'''
try:
  dic = dict(item.split(":") for item in input("Nhập key:value : ").strip().split())
  dic = {k: int(v) for k,v in dic.items()}
  k = int(input("Nhập số nguyên k: "))
  new_dic = {key: val for key, val in dic.items() if val > k}
  print(new_dic)
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```
