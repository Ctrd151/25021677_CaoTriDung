```python
'''W7A16: Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu mỗi phần tử
x trong mảng xuất hiện đúng x lần. Kiểm tra xem mảng có phải là mảng đẹp
hay không. Nếu không hãy xóa đi y số (y là nhỏ nhất có thể) để mảng còn lại thành
mảng đẹp'''
try:
  n = list(map(int,input("Nhập dãy số: ").strip().split()))
  a = set(n)
  flag = True
  count_del = 0
  for i in a:
    count = 0
    for j in n:
      if j == i:
        count += 1
    if count != i:
      if count > i:
        count_del += count - i
      if count < i:
        count_del += count
      flag = False
  if flag:
    print("YES")
  else:
    print("NO " + str(count_del))
except ValueError:
  print("Đầu vào không hợp lệ!!!")
```

```python
'''W7A17: Tìm phần tử xuất hiện nhiều hơn n/2 lần trong mảng (nếu có) với độ phức tạp thời
gian là O(n), độ phức tạp bộ nhớ là O(1), - Boyer–Moore. Nếu không có phần tử
nào thỏa mãn thì không in gì'''
n = list(map(int,input().split()))
dic = {}
for i in n:
  dic[i] = dic.get(i,0) + 1
for k,v in dic.items():
  if v > (len(n)//2):
    print(k)
```

```python
'''W7A17: Tìm phần tử xuất hiện nhiều hơn n/2 lần trong mảng (nếu có) với độ phức tạp thời
gian là O(n), độ phức tạp bộ nhớ là O(1), - Boyer–Moore. Nếu không có phần tử
nào thỏa mãn thì không in gì'''
a = list(map(int, input().split()))

# Bước 1: tìm ứng viên
candidate = None
count = 0

for x in a:
    if count == 0:
        candidate = x
        count = 1
    elif x == candidate:
        count += 1
    else:
        count -= 1

# Bước 2: kiểm tra lại
if candidate is not None and a.count(candidate) > len(a)//2:
    print(candidate)

```

```python
'''W7A18: Sắp xếp có điều kiện: Số chẵn đứng trước, số lẻ đứng sau, chẵn tăng dần, lẻ
giảm dần'''
a = list(map(int,input().split()))
def bubble_sort(n):
  for i in range(len(n)):
    for j in range(len(n)-i-1):
      if n[j] > n[j + 1]:
        n[j], n[j+1] = n[j+1], n[j]
  return n
odd, even =[], []
a = bubble_sort(a)
for i in a:
  if i % 2 == 0:
    even.append(i)
  else:
    odd = [i] + odd
print(*(even + odd))
```

```python
'''W7A19: Merge 2 mảng đã sắp xếp (có thể tăng dần, có thể giảm dần) mà không dùng
bất cứ thuật toán sắp xếp nào thành một mảng cũng có cùng thứ tự sắp xếp như 2
mảng ban đầu'''
a = list(map(int, input().split()))
b = list(map(int, input().split()))

# xác định chiều sắp xếp (giả sử a và b cùng chiều)
inc = a[0] <= a[-1]

i = j = 0
c = []

while i < len(a) and j < len(b):
    if inc:  # tăng dần
        if a[i] <= b[j]:
            c.append(a[i])
            i += 1
        else:
            c.append(b[j])
            j += 1
    else:    # giảm dần
        if a[i] >= b[j]:
            c.append(a[i])
            i += 1
        else:
            c.append(b[j])
            j += 1

# thêm phần còn lại
c.extend(a[i:])
c.extend(b[j:])

print(*c)
```

```python
'''W7A22: Cho N tên sinh viên gồm cả họ và tên, tìm xem tên nào hay gặp nhất. Nếu có
nhiều tên cùng số lần gặp thì lấy tên có thứ tự từ điển nhỏ nhất'''
N = int(input())
name = []
for i in range(N):
  ten = list(map(str, input().split()))
  name.append(ten)
dic = {}
for i in range(N):
  dic[name[i][-1]] = dic.get(name[i][-1],0) + 1
maxk = max(dic.values())
res = min(k for k, v in dic.items() if v == maxk)
print(res)
```

```python
'''W7A23: Đếm số lượng tên trong xâu kí tự. Tên được định nghĩa là xâu có độ dài >=2 và
có chữ cái đầu viết hoa, các chữ cái khác viết thường, chỉ có kí tự cái'''
n = input().split()
count = 0

for i in n:
    if (
        len(i) >= 2 and
        i.isalpha() and
        i[0].isupper() and
        i[1:].islower()
    ):
        count += 1

print(count)
```

```python
'''W7A26: Tách các số có trong một đoạn văn bản, tính tổng và tính trung bình các số đó. '''
s = input()
lst = []
num = ""

for ch in s:
    if ch.isdigit():
        num += ch
    else:
        if num:
            lst.append(int(num))
            num = ""

if num:  # nếu số kết thúc ở cuối chuỗi
    lst.append(int(num))

if lst:
    print(sum(lst), sum(lst)/len(lst))
else:
    print(0, 0)
```

```python
'''W7A27: Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra xem dãy ngoặc đã cho có hợp lệ hay
không'''
s = input().strip()
stack = []
pair = {')': '(', '}': '{', ']': '['}

ok = True
for ch in s:
    if ch in '({[':
        stack.append(ch)
    elif ch in ')}]':
        if not stack or stack[-1] != pair[ch]:
            ok = False
            break
        stack.pop()

if ok and not stack:
    print("YES")
else:
    print("NO")
```

```python
'''W7A28: Bài toán Tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên nào <= N (N
là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ dài thì lấy dãy nhỏ
nhất (gặp đầu tiên từ trái qua)'''
N = int(input())
def hop_so(a):
  if a <= 3:
    return False
  for i in range(2, int(a**0.5)+1):
    if a % i == 0:
      return True
  return False
start_num = None
max_len = 0
current_len = 0
for i in range(4, N):
  if hop_so(i):
    current_len += 1
    if current_len == 1:
      temp_start = i
  else:
    if max_len < current_len:
      max_len = current_len
      start_num = temp_start
    current_len = 0
if current_len > max_len:
    max_len = current_len
    start_num = temp_start
lst = [x for x in range(start_num, start_num + max_len)]
print(*lst)
```

```python
'''W7A29: Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ bàn
phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau'''
lst = list(map(int,input().split()))
lstm = []
lstcr = [lst[0]]
for i in range(1,len(lst)):
  if lst[i] > lst[i-1]:
    lstcr.append(lst[i])
  else:
    if len(lstcr) > len(lstm):
      lstm = lstcr
    lstcr = [lst[i]]

if len(lstcr) > len(lstm):
      lstm = lstcr
print(*lstm)
```

```python
'''W7A30: Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập vào từ
bàn phím'''
lst = list(map(int,input().split()))
lstm = []
for i in range(len(lst)):
  lstcr = [lst[i]]
  temp = lst[i]
  for j in range(i+1, len(lst)):
    if lst[j] >= temp:
      lstcr.append(lst[j])
      temp = lst[j]
  if len(lstcr) >= len(lstm):
    lstm = lstcr
    lstcr = []
print(*lstm)
```
